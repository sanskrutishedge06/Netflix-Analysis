```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df['country'] = df['country'].fillna('United States')
df['cast'] = df['cast'].fillna('No Data')
df['director'] = df['director'].fillna('No Data')
df.dropna(subset=['date_added', 'rating'], inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
```

```python
# A. The Content Mix (Donut Chart)
plt.figure(figsize=(8, 8))
type_counts = df['type'].value_counts()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140,
        colors=['#E50914', '#221F1F'], wedgeprops={'width': 0.4})
plt.title('Content Mix: Movies vs TV Shows')
plt.show()
```

```python
# B. Growth Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df['year_added'].value_counts().sort_index(), color='#E50914', linewidth=3)
plt.title('Content Added to Netflix Over the Years')
plt.ylabel('Count of Titles')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# C. Top 10 Countries producing content
plt.figure(figsize=(10, 6))
top_countries = df['country'].str.split(', ').explode().value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Reds_r')
plt.title('Top 10 Global Content Hubs')
plt.show()
```
